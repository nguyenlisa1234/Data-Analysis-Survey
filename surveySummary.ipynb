# **Data Analysis of the 2023 Stack Overflow Developer Survey**

In May 2023 over 90,000 developers responded to an annual survey about how they learn and level up, which tools they're using, and which ones they want.

In this analysis, the information is broken up into 4 categories comprising of multi-question analysis:
* Developer Profile
* Employment
* Education
* New Technology
* AI Technology 

This data set was analyzed to display percent values, using `.value_counts()` and displayed visually through charts using `matploylib`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('survey_results_public.csv') # will have to trace path of file if notebook is not in same folder
```

# **Developer Profile**
Containing age, industry, country, work experience, currency/annual compensation, management position, and life outside of work.

## What is your age?

```python
%matplotlib inline

ax = df['Age'].value_counts().sort_values(ascending=True).plot.barh(color='pink')

plt.title('Age')
plt.xlabel('Count')
plt.ylabel('Age Range')
 

total = len(df['Age'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.show()
```

## What industry is the company you work for in?

```python
ax = df['Industry'].value_counts().sort_values(ascending=True).plot.barh(color='pink')

plt.title('Industry')
plt.xlabel('Count')
plt.ylabel('Industries')

total = len(df['Industry'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.show()
```

## Where do you live?

```python
df.Country.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
```

## How many years of working experience do you have?

```python
df.WorkExp.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
```

## Which currency do you use day-to-day?  `normalize=True` creates a percentage output

## What is your current total annual compensation (salary, bonuses, and perks, before taxes and deductions)? 

```python
df.CompTotal.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
```
